```python
import re
def remove_html_tags(text):
    if isinstance(text, str):
        pattern = re.compile('<.*?>')
        return pattern.sub(r'', text)
    else:
        return text
```

```python
training_data['article'] = training_data['article'].apply(remove_html_tags)
```

```python
training_data['highlights'] = training_data['highlights'].apply(remove_html_tags)
```

```python
import string
string.punctuation
exclude = string.punctuation
def remove_punc(text):
    if isinstance(text, str):
        return ''.join(ch for ch in text if ch not in exclude)
    else:
        return str(text)  # Convert non-string data to string
```

```python
training_data['article'] = training_data['article'].apply(remove_punc)
```

```python
training_data['highlights'] = training_data['highlights'].apply(remove_punc)
```
